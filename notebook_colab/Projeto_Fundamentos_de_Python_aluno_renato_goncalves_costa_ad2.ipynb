### **Introdução**

Este projeto em Python aborda a análise de duas variáveis, x e y, relacionadas à quantidade de propagandas e vendas realizadas pelo setor de marketing da empresa XPTO. Dividido em três partes, busca-se calcular as métricas de estatística descritiva para ambas as variáveis. A primeira parte enfoca o cálculo das métricas para a quantidade de propagandas (variável x), enquanto a segunda parte se concentra nas métricas para a quantidade de vendas (variável y). Na terceira parte, são abordadas as métricas de associação e um modelo preditivo, especificamente a regressão linear simples. Com uma abordagem didática, o projeto visa oferecer métricas sobre a associação entre a quantidade de propagandas e as vendas e uma possível correlação. Este projeto foi desenvolvido levando em consideração algumas regras de negócio importantes. Uma delas é o uso do coeficiente de Pearson como medida de associação, o qual requer que a quantidade de elementos nas variáveis x e y seja a mesma, formando pares correspondentes. Essa condição é essencial para garantir a precisão das métricas de estatística. Ao inserir a quantidade de elementos nas variáveis x e y, siga as orientações  fornecidas para obter resultados adequados.

### **Orientações gerais ( regra de negócio da empresa XPTO)**

### Algoritmo para calcular a estatistica descritiva para uma amostra de dados (X) referente a quantidade de propagandas realizada pela empresa XPTO.

```python
#Variáveis
variancia_amostral = []
lista_x = []
outliers_superior = []
outliers_inferior = []
maior_valor = 0.0
valor_moda = 0.0
contador = 0
media_x = 0.0
mediana_par = 0.0
# O try e except, realiza o tratamento caso o usuário não atenda as regras de negócio da empresa XPTO
try:
  #Entrada de dados
  qte_dados_x = int(input('Digite a  quantidade de elementos na variável X: '))

  #Verifica se a quantidade de valores inseridos pelo usuário é par.
  if qte_dados_x % 2 == 0:
    while contador < qte_dados_x:
      dados_x = float(input('Digite os elementos para variável x : '))

      # Cálculo da média
      media_x += (dados_x)/qte_dados_x
      contador+=1

      #Populando a lista_salarios com os valores inseridos pelo usuário no input
      lista_x.append(dados_x)

      #Ordenando a lista em ordem crescente com seus respectivos salários base
      lista_x.sort()

      #Calcula a mediana quando a  quantidade de elementos é ímpar
      if len(lista_x) % 2 != 0:
       mediana_impar = (len(lista_x) // 2)

      #Calcula a mediana quando a quantidade de elementos é par
      else:
        mediana = (len(lista_x) // 2)
        mediana1 = mediana -1
        mediana2 = mediana1 + 1
        mediana_par = ((lista_x[mediana1] + lista_x[mediana2])/2)

      #Calcula a moda
      for indice in lista_x:
        contador_moda = lista_x.count(indice)
      if contador_moda > maior_valor:
        maior_valor = contador_moda
        valor_moda = indice
      else:
        valor_moda = 0.0

    #Soma dos quadrados das diferenças
    soma_quadrados = sum(( index - media_x)**2 for index in lista_x)

    #calcular a variância amostral (σ2)
    variancia_amostral = (soma_quadrados /(qte_dados_x - 1))

    #Calcular o Desvio Padrão
    desvio_padrao = ((variancia_amostral)**(1/2))

    #Amplitude total
    amplitude_total = max(lista_x) - min(lista_x)

  #Calcula o quartil 1 e quartil 3
  posicao_decimal = qte_dados_x / 4
  if  posicao_decimal.is_integer():
    q1 = ((lista_x[int(posicao_decimal)- 1]) + (lista_x[int(posicao_decimal)])) / 2
  else:
    posicao_inteira = int(posicao_decimal)
    q1 = lista_x[posicao_inteira]

  posicao_decimal =  qte_dados_x *(3 / 4)
  if posicao_decimal.is_integer():
    q3 = (lista_x[int(posicao_decimal) - 1] + lista_x[int(posicao_decimal)]) / 2
  else:
    posicao_inteira = int(posicao_decimal)
    q3 = lista_x[posicao_inteira]

  # Calcula o intervalo interquartil (IQR)
  iqr = (q3 - q1)

  # Calcula os limites inferior e superior do boxplot
  limite_inferior = q1 - 1.5 * iqr
  limite_superior = q3 + 1.5 * iqr

  #Identificação de possíveis outliers na amostra
  for outliers in lista_x:
    if outliers < limite_inferior:
      outliers_inferior.append(outliers)
    elif outliers > limite_superior:
      outliers_superior.append(outliers)
    else:
      pass
  #Cálculo da mediana
  if contador % 2 == 0:
   mediana_par
  elif contador % 2 == 0:
    mediana_impar
  else:
    pass
  print('DADOS DA AMOSTRA DA VARIÁVEL X')
  #Imprimir a mediana
  print(f'A mediana: {mediana_par:.2f}')
  #Imprime a moda
  print(f'Moda: {valor_moda}')
  #Imprime a lista com todos os salários base
  print(f'Dados  inseridos pelo usuário : {lista_x}')
  #Imprime a média salarial
  print(f'Média: {media_x:.2f}')
  #Imprime a variância amostral (σ2)
  print(f'variância amostral: {variancia_amostral:.2f}')
  #Imprime o desvio padrão
  print(f'O desvio padrão da amostra: {desvio_padrao:.2f}')
  #Imprime o valor da amplitude total
  print(f'Amplitude total: {amplitude_total:.2f}')
  #Imprime o quartil 1 (Q1)
  print(f'Primeiro quartil (Q1): {q1:.2f}')
  #Imprime o quartil 3 (Q3)
  print(f'Terceiro quartil (Q3): {q3:.2f}')
  #Imprime a amplitude interquartílica (AIQ);
  print(f'Amplitude interquartílica(AIQ): {iqr:.2f}')
  #Imprime o Limite inferior (LI do boxplot);
  print(f'Limite inferior(LI): {limite_inferior:.2f}')
  #imprime o Limite superior (LS do boxplot);
  print(f'Limite superior(LS): {limite_superior:.2f}')
  print(f'outliers inferior: {outliers_inferior}')
  print(f'outliers superior: {outliers_superior}')

#O except envia uma mensagem de erro caso o usuário não atenda a regra de negócio (pré-requisto)
except:
  print('''Se você está lendo essa mensagem é porque não atendeu a regra de negócio do algoritmo. Veja as intruções abaixo:
          Regra de negócio 1) Para obter resultados precisos, certifique-se de que a quantidade de elementos nas variáveis X e Y seja a mesma.
          Regra de negócio 2) A quantidade de elementos nas variáveis X e Y deve ser par.
          Regra de negócio 3) Não utilizar caracteres especiais, textos ou valores decimais para definir a quantidade de elementos nas variáveis X e Y.
          Tente novamente! :)''')
```

### Medidas de Associação

* Coeficiente de Correlação de Pearson ( r );
* Coeficiente de determinação (r²).


### MODELO PREDITIVO: REGRESSÃO LINEAR SIMPLES

* Coeficiente a ou alpha da equação de regressão linear simples para as duas variáveis;
* Coeficiente b ou beta da equação de regressão linear simples para as duas variáveis escolhidas pelo usuário.



```python

#Somatoria dos elementos da amostra da variável X (lista_x)
somatoria_lista_x = sum(lista_x)

#Somatoria dos elementos da amostra da variável Y (lista_y)
somatoria_lista_y = sum(lista_y)

# elevar os elementos da amostra da variável X ao quadrado (lista_x)
x_quadrado = []
for i in lista_x:
 x_quadrado.append(i**2)

# elevar os elementos da amostra da variável Y ao quadrado (lista_y)
y_quadrado = []
for i in lista_y:
  y_quadrado.append(i**2)

#somatoria dos elementos da amostra da variável x_quadrado
somatoria_x_quadrado = sum(x_quadrado)

#somatoria dos quadrados dos elementos da amostra da variável y_quadrado
somatoria_y_quadrado = sum(y_quadrado)

#multiplicação das amostra X e Y (Lista_x * Lista_y)
produto_mult_xy = []
for indice in range(len(lista_x)):
  produto_mult_xy.append(lista_x[indice] * lista_y[indice])

#Somatoria da multiplicação das amostras X e Y
somatoria_mult_xy = sum(produto_mult_xy)

#Calculo do numerador do coeficiente de Pearson (Covariância (XY))
coeficiente_pearson_numerador = somatoria_mult_xy - ((somatoria_lista_x * somatoria_lista_y) / len(lista_x))

#Calculo do denominador do coeficiente de Pearson
coeficiente_pearson_denominador_01_x = somatoria_x_quadrado - ((somatoria_lista_x)**2)/len(lista_x)
coeficiente_pearson_denominador_02_y = somatoria_y_quadrado - ((somatoria_lista_y)**2)/len(lista_y)
coeficiente_pearson_denominador = (coeficiente_pearson_denominador_01_x * coeficiente_pearson_denominador_02_y)**(0.5)

#Cálculo coeficiente de pearson (r)
coeficiente_pearson = (coeficiente_pearson_numerador / coeficiente_pearson_denominador)

#Coeficiente de determinação ( r ao quadrado)
coeficiente_pearson_quadrado = (coeficiente_pearson)**2
#Coeficiente de deterrminação em porcentagem %
coeficiente_pearson_percent = (coeficiente_pearson_quadrado * 100)

#Cálculo do coeficiente angular β da equação de regressão linear simples
numerador_beta = somatoria_mult_xy - ((somatoria_lista_x * somatoria_lista_y)/len(lista_x))
denominador_beta = somatoria_x_quadrado - (((somatoria_lista_x)**(2)) /len(lista_x))
beta_coeficiente_angular = (numerador_beta / denominador_beta)

#Cálculo do coeficiente linearα ou alpha da equação de regressão linear simples
alpha_coeficiente_linear = media_y - (media_x * beta_coeficiente_angular)

print(f'coeficiente de pearson: {coeficiente_pearson:.4f}')
print(f'coeficiente de determinação: {coeficiente_pearson_quadrado:.4f}')
print(f'coeficiente de determinação em percentual: {coeficiente_pearson_percent:.2f}%')
print(f'Coeficiente alpha: {alpha_coeficiente_linear:.2f}')
print(f'coeficiente beta: {beta_coeficiente_angular:.2f}')
```
